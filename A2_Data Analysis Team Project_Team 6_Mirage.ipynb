```python
## INITIAL IMPORT AND CREATION OF THE DATASETS#################################

"""
Three dataframe will be accessible: data (with all the original datas of the whole world), 
data_mirage (with the original datas of only our region), and data_mirage_filled 
(with our best guesses on the missing values for the analyzed region.

In order to simplify the reading for the final user, and the handling for the coders, the columns have been renominated 
with more concise wordings.

To compare the values across different countries the two columns "AIDS estimated deaths (UNAIDS estimates)" 
and "Reported cases of malaria" have been transformed into ratios per 100k people (divided by the relative 
population and multiplied by 100,000).

"""

# HERE WE IMPORT THE MAIN PACKAGES NEEDED---------------------------------------

# importing packages
import pandas as pd                  # data science essentials
import matplotlib.pyplot as plt      # data visualization essentials
import seaborn as sns                # enhanced data visualization
from pandas.api.types import is_numeric_dtype
import numpy as np                   # apply in creat scatter plot

# HERE WE IMPORT THE EXCEL IN 4 DIFFERENT DATAFRAME-----------------------------

# storing the path to the dataset
file = "Final_Project_Dataset.xlsx" 

# reading the three sheets of the file into Python through pandas
data = pd.read_excel(io = file,
                   sheet_name = 'Data',
                   header = 0)

series_metadata = pd.read_excel(io = file,
                   sheet_name = 'Series - Metadata',
                   header = 0)

country_metadata= pd.read_excel(io = file,
                   sheet_name = 'Country - Metadata',
                   header = 0)

# renaming of the columns to match "country_metadata" sheet with "data" sheet
country_metadata = country_metadata.rename(columns={'Code': 'code', 
                                                    'Long Name':'long_name',
                                                    'Income Group': 'income_group'})

# renaming of the columns for a better readibility in "data" sheet
data.columns = ('code','name','region','cool_name','aids_deaths_100k','enrollment',
              'fertility','antiretroviral_coverage','birth_skilled','co2_capita',
              'female_contributing','male_contributing','member_contributing',
              'female_empl','male_empl','pop_empl','energy_per_gdp','fertiliy',
              'gdp_employed','gdp_energy','gni_capita','measles_imm','sanitation_access',
              'water_access','tubercolosis_100k','income_lower20','internet_100',
              'life_expectation','literacy_rate','maternal_mortality','mobile_100',
              'infant_mortality_1000','oda_capita','population','poverty_gap',''
              'prenatal_care','hiv','undernourishment','primary_completion',
              'female_parliament','malaria_100k','school_enroll','self_employment',
              'trade','tubercolosis_death')
     
# here we drop the row with the world's data
data = data[data.code != "WLD"]
     
# dividing pure numbers by population to have more consistent values across countries
data['aids_deaths_100k'] = (data['aids_deaths_100k']/data['population'])*100000
data['malaria_100k'] = (data['malaria_100k']/data['population'])*100000


# adding a column to separate Mirage from the rest of the world
data['is_mirage'] = np.where(data['cool_name']=='Mirage', 'mirage', 'rest of the world')


# merging the income group on country_metadata in data
data = data.merge(country_metadata.iloc[ : ,0:3], on = 'code', how='left')

# filtering team cool_name "Mirage"
data_mirage = data.loc[ : , :][data.loc[ : , 'cool_name'] == "Mirage"]
data_mirage = data_mirage.drop(columns = ['cool_name','long_name','is_mirage'])
data = data.drop(columns = ['cool_name','long_name'])
```

```python
# HERE WE FIND THE MISSING VALUES###############################################

# Check the null values per columns
print("""
Columns with null values:
""")
columns_missing = data_mirage.isnull().sum(axis=0)[ data_mirage.isnull().sum(axis=0) > 0].sort_values(ascending = False)
print(columns_missing)

# Check the null values as a percentage of total observations per columns 
print("""
In percentage:
""")
print(data_mirage.isnull().mean(axis = 0).round(decimals = 2)[ data_mirage.isnull().sum(axis=0)> 0].sort_values(ascending = False))

print(f"\n{'-' * 80}")
```

In order to fill in the dataset with the missing values, we decided not to add external numbers in order to maintain the consistency with the rest of the dataset with one exception: "Proportion of seats held by women in national parliaments (%)" (it is easily traceable without further issues).<br>

The strategy consisted in analyzing the most promising of two paths: <br>
1. averaging <strong> OR </strong> <br>
2. linear regression<br>

To decide if an average could have been a good fit for the missing value we went through different perspective: 
- the region itself compared with the rest of the world  <strong> OR </strong>
- the various income subgroups compared with each others in the region and in the world itself.<br>

If the box-plot had a population big enough to give solidity to the analysis and a range of values low enough to suggest that a single, averaged value could represent the subgroup, we deem the median an appropriate guess.<br>

For the linear regression, we started from a correlation matrix between all the columns. <br>
We then followed the analysis only on the strongest correlations (correlation > 0.8) and then extrapolated the linear regression between the values.<br>
Every regression has been analyzed in the light of the following conditions:<br>
1. Enough values to have statistical significance <br>
2. Error evenly distributed around the linear regression and no trends<br>
3. Reasonableness of the correlation (even though correlation doesn't mean causation, the relation between the variables should be reasonable and be argued in words)<br>
4. If one or more of these conditions doesn't subsist, the linear regression may be used for a sub-population (based on the  "Income Group") or for a particular subset of values.<br>

If both the paths resulted in a reasonable value, an average of the two values is taken.

```python
# CORRELATION RESULTS

# HERE WE CREATE THE HEATHMAP -------------------------------------------------

# creating a fancy heatmap and a matrix of correlation
plt.figure(figsize=(20,15))
print(sns.heatmap(data = data.corr(),
                  cmap = "YlOrRd",
                  square = True,
                  linecolor = 'white',
                  linewidths = 0.1))

# title and displaying the plot
plt.title("""
Linear Correlation Heatmap WORLDWIDE
""")

# function to show plot 
plt.show()




# HERE WE FIND AND PRINT THE STRONGEST CORRELATION ( > 0.8) --------------------

# we create a correlation matrix through the .corr() method
corr_matrix = data.corr().round(decimals = 2)

# we create an empty list that will be fulfilled in the next for cycle and will be used for the regressions
corr_list=[]

# a simple counter to show how many "strong" correlation the loop found
corr_num=0

# for loop to find the stronges correlations: it evaluates every rows and column of the correlation matrix
for ro in corr_matrix:
    # in corr_matrix rows and columns have the same index, we can iterate the same for cycle with different names
    for col in corr_matrix:
        
        # extrapolate the value of each row-column
        val = (corr_matrix.loc[ro , col])
        
        # we defined as "strong" correlation a corr > 0.8
        if abs(val) > 0.80 and col != ro:
            
            # counter + 1
            corr_num +=1
            
            # user friendly prints (they will be hidden in the report)
            # print(f"""strong correlation ({corr_matrix.loc[ro , col]})
#between {ro} and {col}""")
            # print("")
            corr_list.append( (ro, col) )

# a final statement to remind the number of correlation founded (it will be hidden in the report)
# print(f"{corr_num} strong correlation founded")




# HERE DEFINE THE FUNCTION FOR THE LINEAR REGRESSIONS AND THE PLOT--------------
# AND USE THEM FOR STRONG CORRELATION

# function to estimate the coefficient of the regression
def estimate_coef(x, y): 
    
    # it's not feasible to drop all the NA values of the main table
    # we can drop the NA only of every specific regression to have enough data
    
    # we create a new dataframe with only the  two columns needed
    data_est = {x : data[x], y: data[y]}
    new_df =pd.DataFrame(data_est)
    
    # drop NA on this new dataframe
    new_df.dropna(axis = 0, inplace = True)

    # we then calculate all the needed factors
    
    # number of observations/points 
    n = np.size(new_df[x]) 
  
    # mean of x and y vector 
    x_mean , y_mean = np.mean(new_df[x]), np.mean(new_df[y]) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(new_df[y]*new_df[x]) - n*y_mean*x_mean 
    SS_xx = np.sum(new_df[x]*new_df[x]) - n*x_mean*x_mean 
  
    # we now can find the coefficients of y = b_0 + b_1*x
    b_1 = SS_xy / SS_xx 
    b_0 = y_mean - b_1*x_mean 
  
    return(b_0, b_1) 


# function to plot the regression and evaluate the outcome of the regression manually
def plot_regression_line(x, y, b): 
    
    # plotting the actual points as scatter plot 
    
    # old code (maintain until the final version if sns.lmplot doesn't work)
    # plt.scatter(data[x], data[y], color= "blue", 
    #           marker = "o", s = 30) 
  
    sns.lmplot(x      = x,  # x-axis feature
           y          = y,  # y-axis feature
           hue        = 'income_group',     # categorical data for subsets
           legend_out = False,    # formats legend if hue != None
           scatter    = True,     # renders a scatter plot
           fit_reg    = False,     # renders a regression line
           aspect     = 3,        # aspect ratio for plot
           data       = data) # DataFrame where features exist


    # predicted response vector 
    y_pred = b[0] + b[1]*data[x] 
  
    # plotting the regression line 
    plt.plot(data[x], y_pred, color = "g") 
  
    # putting labels and title 
    plt.xlabel(x) 
    plt.ylabel(y) 
    plt.title(f'Regression of {x} with {y}')
  
    # function to show plot 
    plt.show() 


# in order to fulfill in a for cycle a dictionary with the coefficients, we need to create an empty dictonary 
regr_coeff = {}

# we launch the regression on every "strong" correlation found with .corr()
for x,y in corr_list:
    
    # we feed with each couple of values the regression function 
    b = estimate_coef(x , y)
 
    # we add new voices in the dictionary: (values x,y):(coefficients b_0, b_1)
    regr_coeff[x,y] = b
    
    
    

#HERE WE FINALIZE THE REGRESSIONS FOR THE MISSING VALUE ----------------------


#Part of code turned off to show all the correlations
#we start again a new for cycle to check the strong correlations we just regressed 
#for x,y in corr_list:
#    
#    #we check if this correlations are related to any missing value
#    if x in columns_missing.index or y in columns_missing.index:
#        
#        #we create a print statement that facilitate the analysis (color formatted, plotted, and with clear statements)
#        if x in columns_missing:
#            print("\033[40m")
#            print("\033[1m")
#            print("\033[97m")
#            print(f"for {x}")
#            print("\033[0m")
#            
#        else:
#            print("\033[40m")
#            print("\033[1m")
#            print("\033[97m")
#            print(f"for {y}")
#            print("\033[0m")
#            
#        print(f"""Y = {y}
#X = {x}
#(B0, B1) = {regr_coeff[x,y]}
#
#
#""") 
#        plot_regression_line(x, y, regr_coeff[x,y])

#the subgroup to be shown in the report
regression_to_show = (['aids_deaths_100k','hiv'],['birth_skilled','infant_mortality_1000'],
                     ['sanitation_access','maternal_mortality'],['water_access','infant_mortality_1000'],
                     ['infant_mortality_1000','literacy_rate'],['primary_completion','infant_mortality_1000'],
                     ['enrollment','school_enroll'])

print("Here the scatterplots we based the regressions on:")    

for x,y, in regression_to_show:
    plot_regression_line(x, y, regr_coeff[x,y])


```

<div style="font-size: 20px"><strong>Part V: Best represents country "on average"</strong></div ><br>

In order to find out the most representative countries, we ranked Mirage's countries based on the top 5 features, giving a score to the first top 3 countries. The following code gave as a result that the most representative country is <em> 'Belarus'</em>, closely followed by Romania and Azerbaijan.
<br><br>

```python
# HERE WE LOOK FOR THE MOST REPRESENTATIVE COUNTRIES FOR THE REGION

# This is the list of the 5 most representative features (+ the code to match with our dataset)
main_features = ('code','member_contributing','energy_per_gdp','fertility','measles_imm','undernourishment')

# we create a new dataframe not to complicate our main dataframes
# given the small amount of data, the inefficiency of dealing with more than one dataframe repays with an increased readability and order
zeros = np.zeros(shape=(len(data_mirage),6))
best_fit_matrix = pd.DataFrame(zeros, columns = main_features, index = data_mirage_filled.index)

# In this for loop we calculate the deviation from the average of every country for every single column
for x in main_features:
    if x == 'code':
        best_fit_matrix['code'] = data_mirage_filled['code']
    else:
        best_fit_matrix[x] = abs((data_mirage_filled[x]-data_mirage_filled[x].median()/data_mirage_filled[x].mean())).round(2)

# we add a column to assing points based on the ranking for each column
best_fit_matrix['counter_rep'] = 0
 
# in this for loop we assign 3 points to every country that is the closest to the average for a single column, 2 points to the second one, and 3 to the third one
for x in main_features:
    if x =='code':
        continue
    rank = best_fit_matrix.sort_values(x,ascending = True)['code'].head(3)
    best_fit_matrix.loc[:,'counter_rep'][best_fit_matrix.loc[:,'code'] == rank.iloc[0]] +=3
    best_fit_matrix.loc[:,'counter_rep'][best_fit_matrix.loc[:,'code'] == rank.iloc[1]] +=2
    best_fit_matrix.loc[:,'counter_rep'][best_fit_matrix.loc[:,'code'] == rank.iloc[2]] +=1
    
# the country with the most points is supposed to be the most representative of the region
print(best_fit_matrix.sort_values('counter_rep',ascending= False).head(3))

# filtering to focus on each relationship with price
sns.pairplot(data = best_fit_matrix,
             x_vars = ['member_contributing','energy_per_gdp','fertility','measles_imm','undernourishment'],
             y_vars ='code',
             hue = 'counter_rep',
             kind = 'scatter',
             palette = 'plasma')


# formatting options and display
plt.tight_layout()
plt.show()
```

<br>
The best represents country on average, which is Belarus may not be able to represent each feature in the same way since the Mirage group is divergent. To illustrate, as shown in member_contributing graph, Belarus has a significantly low rate among Mirage with the fact that the political situation restricted them to have purely free self-employment. However, for the other 4 features, Belarus well represents the group. This is potentially caused by the majority of countries in the Mirage group is in the upper-middle-income group and 7 out of 14 countries, used to be part of the Soviet Union so they share some similar factors.
